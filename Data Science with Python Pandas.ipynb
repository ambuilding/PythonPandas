## Exploring Series

```python
import pandas as pd
import numpy as np
```

```python
# Create a series
pd.Series(np.random.randn(5), index=['a', 'b', 'c', 'd', 'e'])
```

```python
# Preserve the series in variable s for later access
s = pd.Series(np.random.randn(5))
print(s)

print("Dataset length:")
print(len(s))
```

```python
# Index
print(s[0])
print("\n")
print(s[:3])
```

## Exploring DataFrames

```python
# Create a dataframe
df = pd.DataFrame(s, columns = ['Column 1'])
df
```

```python
# Can access columns by name
df['Column 1']
```

```python
# Easy to add columns
df['Column 2'] = df['Column 1'] * 4
df
```

```python
# Other manipulation, like sorting -- if you want to preserve, set equal to a var
df.sort_values(by = 'Column 2')
```

```python
# Boolean indexing
df[df['Column 2'] <= 3]
```

```python
df.apply(lambda x: min(x) + max(x))
np.mean(df)
```
